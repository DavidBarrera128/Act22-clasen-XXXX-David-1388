```python
class Ingeniero:
    def __init__(self, nombre, especialidad, experiencia):
        self.nombre = nombre
        self.especialidad = especialidad
        self.experiencia = experiencia  # Uso de int para los años
        self.trabajando = False

    def asignar_proyecto(self, proyecto):
        self.trabajando = True
        print(f"El ingeniero {self.nombre} ha iniciado el proyecto: {proyecto}.")

    def mostrar_perfil(self):
        print(f"Ingeniero: {self.nombre} | Especialidad: {self.especialidad} | Experiencia: {self.experiencia} años")

# --- Captura de datos para el Ingeniero 1 ---
print(">>> Registro del Primer Ingeniero")
nom1 = input("Nombre: ")
esp1 = input("Especialidad: ")
exp1 = int(input("Años de experiencia: ")) # Conversión a int
ing1 = Ingeniero(nom1, esp1, exp1)

# --- Captura de datos para el Ingeniero 2 ---
print("\n>>> Registro del Segundo Ingeniero")
nom2 = input("Nombre: ")
esp2 = input("Especialidad: ")
exp2 = int(input("Años de experiencia: ")) # Conversión a int
ing2 = Ingeniero(nom2, esp2, exp2)

# --- Ejecución y resultados ---
print("\n" + "="*30)
print("LISTADO DE PERSONAL")
print("="*30)

print("Ingeniero 1:")
ing1.mostrar_perfil()
ing1.asignar_proyecto("Desarrollo de Software")

print("\nIngeniero 2:")
ing2.mostrar_perfil()
ing2.asignar_proyecto("Mantenimiento de Hardware")

print("\n" + "-"*30)
print("Programa terminado")
print("Creador del software por David Barrera NL = 8")
```
